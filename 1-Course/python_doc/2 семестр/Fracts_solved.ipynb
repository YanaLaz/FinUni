```python
class Fract:
    
    def __init__(self, numer, denomin):
        if denomin == 0:
            raise ZeroDivisionError
        self.num = numer
        self.den = denomin
        if type(self.num) == float or type(self.den) == float:
            self.__float_to_int()
        self.__gcd = Fract.gcd(self.num, self.den)
        self.num = self.num // self.__gcd
        self.den = self.den // self.__gcd
    
    def __float_to_int(self):
        res1 = Fract.__float_to_fract(self.num)
        res2 = Fract.__float_to_fract(self.den)
        res3 = res1 / res2
        self.num = res3.num
        self.den = res3.den
        
    @staticmethod
    def __float_to_fract(other):
        other_den = 1
        while True:
            if int(other) == other:
                break
            other_den *= 10
            other *= 10
        return Fract(int(other), int(other_den))
    
    def __float__(self):
        return self.num/self.den
    
    def __int__(self):
        return self.num//self.den
    
    def __tuple__(self):
        pass
    
    def __list__(self):
        pass
    
    def __summ(self, other):
        if type(other) == int:
            return Fract(self.num + other * self.den, self.den)
        elif type(other) == float:
            float_fract = Fract.__float_to_fract(other)
            return Fract(int(self.num*float_fract.den + float_fract.num*self.den), int(self.den*float_fract.den))
        elif isinstance(other, Fract):
            return Fract(self.num*other.den + other.num*self.den, self.den*other.den)
        else:
            raise TypeError
    
    def __multy(self, other):
        if type(other) == int:
            return Fract(self.num*other, self.den)
        elif type(other) == float:
            float_fract = Fract.__float_to_fract(other)
            return Fract(int(self.num*float_fract.num), int(self.den*float_fract.den))
        elif isinstance(other, Fract):
            return Fract(self.num*other.num, self.den*other.den)
        else:
            raise TypeError
            
    #Сложение
    def __add__(self, other):
        return self.__summ(other)
            
    
    def __radd__(self, other):
        return self.__summ(other)
    
    #Умножение
    def __mul__(self, other):
        return self.__multy(other)
    
    def __rmul__(self, other):
        return self.__multy(other)
    
    @staticmethod
    def gcd(a, b):
        if b == 0:
            return a
        else:
            return Fract.gcd(b, a%b)
        
    #Вычитание
    def __sub__(self, other):
        if type(other) == int:
            return Fract(self.num - other * self.den, self.den)
        elif type(other) == float:
            float_fract = Fract.__float_to_fract(other)
            return Fract(int(self.num*float_fract.den - float_fract.num*self.den), int(self.den*float_fract.den))
        elif isinstance(other, Fract):
            return Fract(self.num*other.den - other.num*self.den, self.den*other.den)
        else:
            raise TypeError
            
    def __rsub__(self, other):
        if type(other) == int:
            return Fract(-self.num + other * self.den, self.den)
        elif type(other) == float:
            float_fract = Fract.__float_to_fract(other)
            return Fract(int(-self.num*float_fract.den + float_fract.num*self.den), int(self.den*float_fract.den))
        elif isinstance(other, Fract):
            return Fract(-self.num*other.den + other.num*self.den, self.den*other.den)
        else:
            raise TypeError
           
    def __tdiv(self, other):
        if type(other) == int:
            return Fract(self.num, self.den*other)
        elif type(other) == float:
            float_fract = Fract.__float_to_fract(other)
            return Fract(int(self.num*float_fract.den), int(self.den*float_fract.num))
        elif isinstance(other, Fract):
            return Fract(self.num*other.den, self.den*other.num)
        else:
            raise TypeError
    
    def __rtdiv(self, other):
        if type(other) == int:
            return Fract(self.den*other, self.num)
        elif type(other) == float:
            float_fract = self.__float_to_fract(other)
            return Fract(int(self.den*float_fract.num), int(self.num*float_fract.den))
        elif isinstance(other, Fract):
            return Fract(self.den*other.num, self.den*other.num)
        else:
            raise TypeError
    
    def __pow__(self, other):
        if isinstance(other, Fract):
            return Fract(self.num**float(other), self.den**float(other))
        else:
            return Fract(self.num**other, self.den**other)
        
    #Деление
    def __truediv__(self, other):
        return self.__tdiv(other)
    
    def __div__(self, other):
        return self.__tdiv(other)
    
    def __rdiv__(self, other):
        return self.__rtdiv(other)
    
    def __rtruediv__(self, other):
        return self.__rtdiv(other)
    
    # <=
    def __le__(self, other):
        return
    
    # >=
    def __ge__(self, other):
        return
    
    # >
    def __gt__(self, other):
        return
    
    # <
    def __lt__(self, other):
        return
    
    # ==
    def __eq__(self, other):
        return
    
    # !=
    def __ne__(self, other):
        return
    
    #математические операции с присваиванием
    def __imul__(self, other):
        pass
    
    def __isub__(self, other):
        pass
    
    def __iadd__(self, other):
        pass
    
    def __idiv__(self, other):
        pass
    
    def __itruediv__(self, other):
        pass
    
    def __pprint(self):
        if self.den == 1 or self.num == 0:
            return f"{self.num}"
        else:
            return f"{self.num}/{self.den}"
        
    def __str__(self):
        return self.__pprint()
    
    def __repr__(self):
        return self.__pprint()
```
