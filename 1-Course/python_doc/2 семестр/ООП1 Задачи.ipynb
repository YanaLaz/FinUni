```python
# задача 10
class Transport():
    def __init__(self, name, koord):
        self.name = name
        self.koord = koord
        
    def info_koord(self):
        return self.koord
        
    def info(self):
        return f"Название: {self.name}, Кооридинаты: {self.koord}"
    
class Plane(Transport):
    def __init__(self, name, max_sp, count, koord):
        super().__init__(name, koord)
        self.max_sp = max_sp
        self.count = count
        
    def info(self):
        return f"Марка: {self.name}, Макс скорость: {self.max_sp}, Кол-во пассажиров: {self.count}, Кооридинаты: {self.koord}"
    
p = Plane('airbus', '1000', '500', '52.12345')
p.info()

class Auto(Transport):
    def __init__(self, name, number, year, koord):
        super().__init__(name, koord)
        self.number = number
        self.year = year
        
    def info(self):
         return f"Марка: {self.name}, Номер: {self.number}, Год: {self.year}, Кооридинаты: {self.koord}"
    
a = Auto('bmv', '224', '2013', '51.123453')
a.info()

class Ship(Transport):
    def __init__(self, name, koord, speed, count, port):
        super().__init__(name, koord)
        self.speed = speed
        self.count = count
        self.port = port
        
    def info(self):
        return f"Название: {self.name}, Кооридинаты: {self.koord}, Скорость: {self.speed}, Кол-во пассажиров: {self.count}, Порт: {self.port}"
        
s = Ship('submarine', '53.16756454', '200', '300', 'Marina')
s.info()

sg = []
sg.append(p)
sg.append(a)
sg.append(s)

for i in sg:
    print(i.info())

print(' ')
print('***Найдём подходящий транспорт***')
    
def find_po_koord(koord_min, koord_max):
    for i in sg:
        if koord_min < i.info_koord() < koord_max:
            print(i.info())

find_po_koord('50.12345', '52.1234')           
```

```python
# задача 12
class Telo():
    def __init__(self, a, b, c):
        self.a =a
        self.b = b
        self.c = c
    
    def info(self):
        print('Figure')
        print('Sides: ' + str(self.a)+ ',' + str(self.b)+ ',' + str(self.c))
        print("Area: " + str(self.area()))
        print("Volume: " + str(self.volume()))
        
    def area(self):
        return 2*(self.a*self.b + self.b*self.c + self.c*self.a)
    
    def volume(self):
        return self.a*self.b*self.c
    
class Parallel(Telo):
    def __init__(self, a, b, c):
        super().__init__(a, b, c)
       
    def info(self):
        print('Parallelepiped')
        print('Sides: ' + str(self.a)+ ',' + str(self.b)+ ',' + str(self.c))
        print("Area: " + str(self.area()))
        print("Volume: " + str(self.volume()))
    
p = Parallel(1,2,3)
#p.info()

class Ball(Telo):
    def __init__(self, r):
        self.r = r
        
    def info(self):
        print('Ball')
        print('Radius: ' + str(self.r))
        print("Area: " + str(self.area()))
        print("Volume: " + str(self.volume()))
        
    def area(self):
        return 4*3.14*self.r*self.r
    
    def volume(self):
        return (4*3.14*self.r**3)/3
    
b = Ball(1)
#b.info()

class Pyramid(Telo):
    def __init__(self, a, b, h, s_bok):
        self.a =a
        self.b = b
        self.h = h
        self.s_bok = s_bok
        
    def info(self):
        print('Piramida')
        print('Sides: ' + str(self.a)+ ',' + str(self.b))
        print('Height: ' + str(self.h))
        print("Area: " + str(self.area()))
        print("Volume: " + str(self.volume()))
        
    def area(self):
        return self.a*self.b+self.s_bok
    
    def volume(self):
        return (self.a*self.b*self.h)/3
    
py = Pyramid(1,4,5,30)

sy = []
sy.append(p)
sy.append(b)
sy.append(py)

for i in sy:
    print(i.info())
    print(' ')
```

```python
#задача 13
import math

class Uravn():
    def __init__(self, a, b):
        self.a = a
        self.b = b
        
    def rez(self):
        print(f'a = {self.a}, b = {self.b}')
        if (self.a == 0 and self.b == 0):
            return("Бесконечное количество решений.")
        if (self.a == 0 and self.b != 0):
            return("Решений нет.")
        if (self.a != 0):
            return 'x = '+str(self.b/self.a)
        
class Lineyn(Uravn):
    def __init__(self,a,b):
        super().__init__(a,b)
        
l = Lineyn(1,5)

class Kvadr(Uravn):
    def __init__(self, a, b, c):
        super().__init__(a, b)
        self.c = c
        
    def rez(self):
        print(f'a = {self.a}, b = {self.b}, c = {self.c}')
        discr = self.b**2 - 4 * self.a * self.c
        if discr > 0:
            x1 = (-self.b + math.sqrt(discr)) / (2 * self.a)
            x2 = (-self.b - math.sqrt(discr)) / (2 * self.a)
            return("x1 = %.2f, x2 = %.2f" % (x1, x2))
        elif discr == 0:
            x = -self.b / (2 * self.a)
            return("x = %.2f" % x)
        else:
            return("Корней нет")
        
k = Kvadr(1,6,4)
            
         
sx = []
sx.append(l)
sx.append(k)

for i in sx:
    print(i.rez())
    print(' ')
```

```python
#задача 15
class Progress():
    def __init__(self, a1, an, n):
        self.a1 = a1
        self.an = an
        self.n = n
        
    def summa(self):
        return 'Сумма прогрессии: ' + str(((self.a1+self.an)*self.n)/2)
    
class Arifm(Progress):
    def __init__(self, a1, an, n):
        super().__init__(a1,an,n)
        
a = Arifm(1,5,5)
a.summa()

class Geometr(Progress):
    def __init__(self, a1,an,n, an1):
        super().__init__(a1,an,n)
        self.an1 = an1
        
    def summa(self):
        q = self.an1/self.an
        return 'Сумма геометрической прогрессии: ' + str((self.a1-self.an*q)/(1-q))    
g = Geometr(1,10,10,2)
g.summa()

sa = []
sa.append(a)
sa.append(g)
for i in sa:
    print(i.summa())
```
