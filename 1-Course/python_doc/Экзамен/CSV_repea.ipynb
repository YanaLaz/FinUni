```python
# 1
# Что дожно произойти. Берем ФИО студента, определяем его ID, дальше по ID в списке всех оценок ищем оценки нашего ученика(ID)
import csv

def stud_id(name):
    try:
        name = name.split(' ')
        file_name = 'students.csv'
        with open(file_name, "r") as f_obj:
            reader = csv.DictReader(f_obj, delimiter=';')
            for row in reader:
                if row['last_name'] == name[0] and row['first_name'] == name[1]:
                    return row['id']
    except Exception as e:
         print('Reading error: {}'.format(str(e)))

def aver(name):
    id_stud = stud_id(name)
    file_name = 'results.csv'
    m = []
    if id_stud:
        try:
            with open(file_name, 'r') as f_obj:
                reader = csv.DictReader(f_obj, delimiter=';')
                for row in reader:
                    if row['student_id'] == id_stud:
                        m  += [int(row['total'])]
                #print(m)
                return (sum(m)/len(m))
        except Exception as e:
            print('Reading error: {}'.format(str(e)))
    else:
        return None
    
aver('Абарыков Андрей')
```

Необходимо реализовать функцию, которая принимает один аргумент - наименование предмета, и один необязательный аргумент boolean=True. Если boolean == True, необходимо вернуть список преподавателей, которые вели данный предмет, иначе необходимо вернуть список групп, у которых был данный предмет

```python
#3
def SubId(sub_name):
    try:
        file_name = 'subjects.csv'
        with open(file_name, "r") as f_obj:
            reader = csv.DictReader(f_obj, delimiter=';')
            for row in reader:
                if row['subject_name'] == sub_name:
                    return row['id']
            else:
                return None
    except Exception as e:
         print('Reading error: {}'.format(str(e)))

def grp_id(gr_id):
    try:
        file_name = 'groups.csv'
        with open(file_name, "r") as f_obj:
            reader = csv.DictReader(f_obj, delimiter=';')
            for row in reader:
                if row['id'] == gr_id:
                    return row['name']
    except Exception as e:
         print('Reading error: {}'.format(str(e)))
            
def grp_name(st_id):
    try:
        file_name = 'students.csv'
        with open(file_name, "r") as f_obj:
            reader = csv.DictReader(f_obj, delimiter=';')
            for row in reader:
                if row['id'] == st_id:
                    #group_name = grp_id(row['group_id'])
                    return grp_id(row['group_id'])
    except Exception as e:
         print('Reading error: {}'.format(str(e)))

def TeachId(teach_id):
    try:
        file_name = 'teachers.csv'
        with open(file_name, "r") as f_obj:
            reader = csv.DictReader(f_obj, delimiter=';')
            for row in reader:
                if row['id'] == teach_id:
                    return row['last_name'],row['first_name'],row['middle_name']
    except Exception as e:
         print('Reading error: {}'.format(str(e)))
            
def lisToS(sub_name,boolean=True):
    sub_id = SubId(sub_name)
    lis = set()
    try:
        file_name = 'results.csv'
        if boolean:
            with open(file_name, 'r') as f:
                reader = csv.DictReader(f, delimiter = ';')
                for row in reader:
                    if row['subject'] == sub_id:
                        teacher_name = ' '.join(TeachId(row['teacher_id']))
                        lis.add(teacher_name)
                return sorted(list(lis))
        else:
            with open(file_name, 'r') as f:
                reader = csv.DictReader(f, delimiter = ';')
                for row in reader:
                    if row['subject'] == sub_id:
                        group_name = grp_name(row['student_id'])
                        lis.add(group_name)
                return sorted(list(lis))
    except Exception as e:
        print(f'Ошибка {str(e)}')
    
print(lisToS('Английский язык',0))
lisToS('Английский язык',1)
```

Реализовать функцию, которая принимает один параметр - наименование предмета. Функция должна вернуть словарь, где ключом является наименование группы, а значением словарь - ключ - оценка по 5-и бальной шкале, значение - количество студентов.

```python
#4
def GroupId(gr_id):
    try:
        file_name = 'groups.csv'
        with open(file_name, "r") as f_obj:
            reader = csv.DictReader(f_obj, delimiter=';')
            for row in reader:
                if row['id'] == gr_id:
                    return row['name']
    except Exception as e:
         print('Reading error: {}'.format(str(e)))
            
def GroupName(st_id):
    try:
        file_name = 'students.csv'
        with open(file_name, "r") as f_obj:
            reader = csv.DictReader(f_obj, delimiter=';')
            for row in reader:
                if row['id'] == st_id:
                    #group_name = grp_id(row['group_id'])
                    return GroupId(row['group_id'])
    except Exception as e:
         print('Reading error: {}'.format(str(e)))

def SubId(sub_name):
    try:
        file_name = 'subjects.csv'
        with open(file_name, "r") as f_obj:
            reader = csv.DictReader(f_obj, delimiter=';')
            for row in reader:
                if row['subject_name'] == sub_name:
                    return row['id']
            else:
                return None
    except Exception as e:
         print('Reading error: {}'.format(str(e)))

def GroupMar(sub_name):
    sub_id = SubId(sub_name)
    groups = {}
    try:
        file_name = 'results.csv'
        with open (file_name, 'r') as f:
            reader = csv.DictReader(f, delimiter=';')
            for row in reader:
                if row['subject'] == sub_id:
                    group_name = GroupName(row['student_id'])
                    groups.setdefault(group_name,{})
                    if int(row['total']) > 85:
                        groups[group_name].setdefault(5,[])
                        groups[group_name][5].append(row['student_id'])
                    elif int(row['total']) > 70:
                        groups[group_name].setdefault(4,[])
                        groups[group_name][4].append(row['student_id'])
                    elif int(row['total']) > 50:
                        groups[group_name].setdefault(3,[])
                        groups[group_name][3].append(row['student_id'])
                    else:
                        groups[group_name].setdefault(2,[])
                        groups[group_name][2].append(row['student_id'])
                    
            for k, v in groups.items():
                for kk, vv in v.items():
                    groups[k][kk] = len(vv)
            
            return groups
    except Exception as e:
        print(f'Ошибка {str(e)}')
        
GroupMar('Английский язык')
res = GroupMar('Английский язык')
print(res,'\n')
try:
    for k in sorted(res.keys()):
        print(k)
        for kk in sorted(res[k].keys()):
            print(f'\t {kk} - {res[k][kk]}')
except:
    print('Ничего важного')
```
