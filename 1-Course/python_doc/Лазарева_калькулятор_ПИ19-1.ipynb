```python
# функция находит неизвестную в первом линейном уравнении
def find_variable(kl):
    for i, e in enumerate(kl[0]):
        if e in symbols:
            global variable
            variable = e
            break    
    return variable
```

```python
# функция выражает значение переменной в первом линейном уравнении
def virazim_var(variable): 
    virazim_varr = kl[0][kl[0].index('=')+1:]
    if virazim_varr[0] == '0':
        virazim_var = (kl[0][kl[0].index(variable)+1:kl[0].index('=')])
        v = []
        for i in virazim_var:
            if i == '-':
                v.append('+')
            elif i == '+':
                v.append('-')
            else:
                v.append(i)
        return v
    else:
        virazim_var = (kl[0][kl[0].index(variable)+1:kl[0].index('=')])
        for i in virazim_var:
            if i == '-':
                virazim_var.insert(virazim_var.index('-'),'+')
                virazim_var.remove('-') 
            elif i == '+':
                virazim_var.insert(virazim_var.index('+'),'-')
                virazim_var.remove('+')
        virazim_varr.extend(virazim_var)
        return virazim_varr
```
