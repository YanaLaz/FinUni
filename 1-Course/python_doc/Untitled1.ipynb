```python
board = [i for i in range(1,10)]

def draw_board(board):
    print("\n",board[0],"|",board[1],"|", board[2])
    print("----------")
    print("\n",board[3],"|",board[4],"|", board[5])
    print("----------")
    print("\n",board[6],"|",board[7],"|", board[8], "\n")

def take_input(player_token): # функция получает и проверяет значение
    valid = False
    while not valid:
        player_answer = input("Куда поставим " + player_token + "? ")
        while not(player_answer.isdigit()):
                player_answer = input("Некорректный ввод. Куда поставим " + player_token+"? ")
        else:
            player_answer = int(player_answer)
        if player_answer >= 1 and player_answer <= 9:
            if (str(board[player_answer-1]) not in "XO"): # проверка на пустую клетку
                board[player_answer-1] = player_token
                valid = True
            else:
                print ("Эта клеточка уже занята")
        else:
            print ("Некорректный ввод. Введите число от 1 до 9 чтобы походить.")

def check_win(board):
    win_value = ((0,1,2),(3,4,5),(6,7,8),(0,3,6),(1,4,7),(2,5,8),(0,4,8),(2,4,6)) # кортеж с выигрышными комбинациями
    for each_coord in win_value:
        if board[each_coord[0]] == board[each_coord[1]] == board[each_coord[2]]:
            return board[each_coord[0]] # если символы во всех трёх заданных клетках равны, возвращаем выигрышный символ
    return False # иначе возвращаем значение False

def main(board):
    counter = 0
    win = False
    while not win:
        draw_board(board)
        if counter % 2 == 0: # если counter чётный, то ход первого игрока(он у нас чётный, потому что counter = 0)
            take_input("X")  # take_input(player_token)
        else:
            take_input("O")
        counter += 1
        winner = check_win(board) # проверяем наличие победителя
        if winner:
            print (winner, 'Выиграл!')
            win = True
            break
        if counter == 9:
            print('Ничья!')
            break
    draw_board(board)

main(board)
```
