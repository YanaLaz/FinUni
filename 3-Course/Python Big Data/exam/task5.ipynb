5. По данным из файла data/meals.json сформировать словарь, в котором по идентификатору блюда можно получить список ингредиентов.

```python
import json
```

```python
with open ('./data/meals.json', 'r', encoding = 'utf-8') as fp:
    data = json.load(fp)
    
data
```

```python
def check_meal(idMeal):
    res = []
    for d in data.get('meals'):
        if d['idMeal'] == idMeal:
            for i in range(1,21):
                if d[f'strIngredient{i}'] != '':
                    res.append(d[f'strIngredient{i}'])
            print(res)
            break
    else:
        raise ValueError
            
check_meal('53050')
```
