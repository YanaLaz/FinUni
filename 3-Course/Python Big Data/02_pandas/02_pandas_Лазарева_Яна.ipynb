5.2 При помощи объединения таблиц и группировок, создайте `DataFrame`, состоящий из трех столбцов: `recipe_id`, `name`, `review_count`. У рецептов, для которых отсутствуют отзывы, в соответствущем столбце должен быть указан 0. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и выведя на экран строку из полученного `DataFrame`, содержащую информацию об этом отзыве.


```python
mr2 = pd.merge(recipes, reviews, how='right', left_on='id', right_on='recipe_id')
mr2.drop(['minutes', 'id', 'minutes', 'contributor_id', 'submitted',
               'n_steps', 'n_ingredients', 'description_length', 'user_id',
               'name_word_count', 'date', 'review', 'rating', 'year_'],  axis=1, inplace=True)
mr2
```

```python
mr3 = pd.merge(mr2, mr2.groupby('recipe_id')['description'].count(), 
                       how='left', left_on='recipe_id', right_on='recipe_id')
mr3.drop_duplicates('name', inplace=True)
mr3.drop('description_x',  axis=1, inplace=True)
mr3.rename(columns={'description_y': 'review_count'}, inplace=True)
mr3.reset_index(drop=True, inplace=True)

mr3 = mr3.reindex(['recipe_id', 'name', 'review_count'], axis=1)

mr3
```
