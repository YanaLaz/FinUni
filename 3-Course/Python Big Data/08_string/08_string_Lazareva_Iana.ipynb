1.2 Напишите функцию `show_info`, которая для рецепта по его `id` создает строку (в смысле объекта python) с описанием следующего вида:

```
"Название"

1. Шаг 1.

----------
#тэг1 #тэг2
```

    
Данные для создания строки получите из файлов `recipes_sample_with_tags_ingredients.csv`, `steps_sample.xml` (__ЛР4__) и `tags_sample.csv` (__ЛР5__). 
Выведите созданную строку на экран.

````python
import requests
from bs4 import BeautifulSoup

with open('../04_data_files/data/steps_sample.xml', 'r', encoding='utf8') as bsf:
    ab = BeautifulSoup(bsf, 'xml')

recipe_steps_dics = {}
for recipe in ab.find_all('recipe'):
    for ids in recipe.find_all('id'):
        recipe_steps_dics[ids.text] = []
        for step in recipe.find_all('step'):
            recipe_steps_dics[ids.text].append(step.text)
````

````python
import json

tags_dict = {}
with open('../05_data_files/tags_sample.json') as jf:
    tags_dict = json.load(jf)
````

````python
recipes_with_steps = pd.read_csv('../05_data_files/recipes_sample_with_tags_ingredients.csv', sep=',', index_col=0)
recipes_with_steps.head()
````

````python
def show_info(recipes_sample, id):
    row = recipes_sample[recipes_sample.id == id]
    res_str = f'"{row.iloc[0][1]}"\n\n'
    steps_lst = recipe_steps_dics[str(id)]
    
    for index, el in enumerate(steps_lst):
        res_str += f'{index+1}. {el}\n'
    
    res_str += '{:-<100}'.format('-')
    res_str += '\n'
    
    tags_lst = tags_dict[str(id)]
    for el in tags_lst:
        res_str += f'#{el} '
        
    return res_str
````

````python
st31 = []
for row in s['review']:
    st = nltk.sent_tokenize(str(row))
    for i in st:
        i = re.sub("[.,:!?]","",i)
        st31.append(nltk.word_tokenize(str(i)))
````

3.2 Посчитайте кол-во уникальных слов среди всех отзывов (без учета регистра и знаков препинания).

````python
pattern = re.compile(r'(^(^[!"#$%&\'()*+\,\-.\/:;<=>?@[\]^_`{|}~]|^[0-9]))')
data = []
for sentences in st31:
    new_words = []
    [[new_words.append(y.lower()) for y in x if pattern.match(y) is None] for x in sentences]
    data.extend(new_words)
````

3.4 Напишите функцию, которая для заданного предложения выводит информацию о частях речи слов, входящих в предложение, в следующем виде:
```
PRP   VBD   DT      NNS     CC   VBD      NNS        RB   
 I  omitted the raspberries and added strawberries instead
``` 
Для определения части речи слова можно воспользоваться `nltk.pos_tag`.

Проверьте работоспособность функции на любом предложении из отзывов.


````python
nltk.download('averaged_perceptron_tagger')
````

````python
def p34(sent):
    str_first, str_second = [],[]
    f = nltk.pos_tag(nltk.word_tokenize(sent))
    for i in f:
        word_len = max(map(len,i))+1
            
        b1 = "{0:^{word_len}}".format(i[1],word_len=word_len)
        str_first.append(b1)

        b2 = "{0:^{word_len}}".format(i[0], word_len=word_len)
        str_second.append(b2)
      
    print(f"{' '.join(str_first)}\n{' '.join(str_second)}\n")
````
